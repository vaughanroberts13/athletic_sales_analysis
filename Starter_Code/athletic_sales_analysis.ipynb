```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales2020 = pd.read_csv('athletic_sales_2020.csv')
sales2021 = pd.read_csv('athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
display(sales2020.head())
display(sales2020.tail())
```

```python
# Display the 2021 sales DataFrame
display(sales2021.head())
display(sales2021.tail())
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales2020.info()
```

```python
# Check the 2021 sales data types.
sales2021.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
salesdata_2020_2021 = pd.concat([sales2020, sales2021], axis='rows', join='inner').reset_index(drop=True)
salesdata_2020_2021
```

```python
# Check if any values are null.
salesdata_2020_2021.isna().sum()
```

```python
# Check the data type of each column
salesdata_2020_2021.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
salesdata_2020_2021['invoice_date'] = pd.to_datetime(salesdata_2020_2021['invoice_date'], errors = 'coerce')
salesdata_2020_2021
```

```python
# Confirm that the "invoice_date" data type has been changed.
salesdata_2020_2021.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_products_sold = salesdata_2020_2021.groupby(['region','state','city'])\
            ['units_sold'].agg(Total_Products_Sold=("sum"))

# Show the top 5 results.
total_products_sold.sort_values(by=(['Total_Products_Sold']), ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_sold = pd.pivot_table(salesdata_2020_2021,
                                     index=['region','state','city'],
                                     values='units_sold',
                                     aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
total_products_sold = total_products_sold.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
total_products_sold.sort_values(by=("Total_Products_Sold"), ascending=False).head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_footwear_sales = pd.pivot_table(filtered_womens_footwear, index=['invoice_date'],
                                            values='total_sales',
                                            aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
womens_footwear_sales_date = womens_footwear_sales.rename(columns={"total_sales": "Total Sales"})


# Show the table.
womens_footwear_sales_date.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_womens_footwear_sales = womens_footwear_sales_date.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_womens_footwear_sales.sort_values(by="Total Sales", ascending=False).head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
daily_womens_footwear_sales = womens_footwear_sales_date.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_womens_footwear_sales.sort_values(by="Total Sales", ascending=False).head(10)
```
